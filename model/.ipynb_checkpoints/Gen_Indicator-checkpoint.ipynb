```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

# Importing dataset of classified irregularities

```python
dataPath = '../classification/'
fileName = 'CGU_data_Class_Final.xlsx'
data = pd.read_excel(dataPath+fileName)
data.head()
```

# Importing dataset of indicators (basis for final)

```python
dataPath = '../data/CGU/'
fileName = 'CGU_data_agg_mun.xlsx'
data_ind = pd.read_excel(dataPath+fileName)
data_ind.head()
```

# Irregularities associated to corruption

```python
ind_corrup = data[['Id_Mun_Sort', 'Id_Irreg','voting_2']].groupby('Id_Mun_Sort').sum().reset_index()
ind_corrup = ind_corrup.rename(index=str, columns={'voting_2': 'n_corrup'})
```

```python
ind_corrup['ind_corrup'] = np.where(ind_corrup['n_corrup']>=1, 1, 0)
```

```python
data_ind_1 = data_ind.merge(ind_corrup, on='Id_Mun_Sort', how='left')
data_ind_1
```

# Irregularities in Education associated to corruption

```python
data_educ = data[data['Orgao_Sup_cod'] == 'EDUC'].reset_index(drop=True)
len(data_educ)
```

```python
ind_corrup_educ = data_educ[['Id_Mun_Sort', 'voting_2']].groupby('Id_Mun_Sort').sum().reset_index()
ind_corrup_educ = ind_corrup_educ.rename(index=str, columns={'voting_2': 'n_corrup_EDUC'})
```

```python
ind_corrup_educ['ind_corrup_EDUC'] = np.where(ind_corrup_educ['n_corrup_EDUC']>=1, 1, 0)
```

```python
len(ind_corrup_educ[ind_corrup_educ['ind_corrup_EDUC']==0])
```

```python
ind_corrup_educ
```

```python
data_ind_2 = data_ind_1.merge(ind_corrup_educ, on='Id_Mun_Sort', how='left')
data_ind_2
```

It is important to note that in our dataset only **1197** municipalities were audited towards Education Ministry (26 had no OS). From these, **16** had OS in Education, but zero irregularities and **109** had irregularities but not classified as corruption. So in total, **125** municipalities had no corruption in Education according to this indicator.

# Severe irregularities in Education associated to corruption

```python
data_educ_grave = data[(data['Orgao_Sup_cod'] == 'EDUC')&(data['Tipo_constatacao'] == 'Falha Grave')].reset_index(drop=True)
len(data_educ_grave)
```

```python
ind_corrup_educ_grave = data_educ_grave[['Id_Mun_Sort', 'voting_2']].groupby('Id_Mun_Sort').sum().reset_index()
ind_corrup_educ_grave = ind_corrup_educ_grave.rename(index=str, columns={'voting_2': 'n_corrup_EDUC_grave'})
```

```python
ind_corrup_educ_grave['ind_corrup_EDUC_grave'] = np.where(ind_corrup_educ_grave['n_corrup_EDUC_grave']>=1, 1, 0)
```

```python
len(ind_corrup_educ_grave[ind_corrup_educ_grave['ind_corrup_EDUC_grave']==0])
```

```python
ind_corrup_educ_grave
```

```python
data_ind_3 = data_ind_2.merge(ind_corrup_educ_grave, on='Id_Mun_Sort', how='left')
data_ind_3
```

From the **1197** municipalities audited towards Education Ministry, only **579** had Severe irregularities and from these, **72** were not classified as corruption. So in total **690** had no corruption in Education according to this indicator.

# Proportion indicators

```python
data_ind_final = data_ind_3.fillna(0)
```

```python
data_ind_final['prop_corrup'] = data_ind_final['n_corrup']/data_ind_final['n_irreg']
data_ind_final['prop_corrup_EDUC'] = data_ind_final['n_corrup_EDUC']/data_ind_final['n_irreg_EDUC']
data_ind_final['prop_corrup_EDUC_grave'] = data_ind_final['n_corrup_EDUC_grave']/data_ind_final['n_irreg_EDUC']
```

# Average by OS indicators

```python
data_ind_final['avg_corrup'] = data_ind_final['n_corrup']/data_ind_final['n_OS']
data_ind_final['avg_corrup_EDUC'] = data_ind_final['n_corrup_EDUC']/data_ind_final['n_OS_EDUC']
data_ind_final['avg_corrup_EDUC_grave'] = data_ind_final['n_corrup_EDUC_grave']/data_ind_final['n_OS_EDUC']
```

```python
data_ind_final = data_ind_final.fillna(0)
```

# Saving to FINAL_DATA folder...

lembrar do indicador de ser auditado pela segunda vez...

```python
dataPath = '../data/FINAL_DATA/'
fileName = 'dataset_CGU.xlsx'
data_ind_final.to_excel(dataPath+fileName)
```
